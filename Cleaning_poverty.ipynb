```python
import pandas as pd
```

```python
poverty = pd.read_csv('./SF/aff_download/ACS_17_5YR_B17001_with_ann.csv')
poverty.head(10)
```

```python
poverty.dtypes
```

```python
poverty = poverty[['GEO.id2', 'HD01_VD01', 'HD01_VD02']]
poverty.head(10)
```

```python
poverty.drop([0,1,2] , inplace=True)
poverty.head(15)
```

```python
poverty.dtypes
```

```python
poverty.to_csv('poverty_clean.csv', index=False)
```
