```python
incomeSubs = income[['NEW YORK', 'Unnamed: 1', 'Unnamed: 19', 'Unnamed: 20']]
incomeSubs.head(25)
```

```python
incomeSubs.rename(columns = {'NEW YORK':'zipcode', 'Unnamed: 1':'buckets', 
                             'Unnamed: 19':'Number of returns', 'Unnamed: 20':'Amount'}, inplace = True) 
incomeSubs.head(20)
```

```python
# indexNames = incomeSubs[ ( incomeSubs['buckets'].isna() &  pd.notna(incomeSubs['zipcode']) ) ].index
indexNames = incomeSubs[ ( pd.notna(incomeSubs['buckets']) |  incomeSubs['zipcode'].isna() ) ].index
indexNames
```

```python
incomeSubs.drop(indexNames , inplace=True)
incomeSubs.head(15)
```

```python
incomeSubs.drop([0,1] , inplace=True)
incomeSubs.head(15)
```

```python
incomeSubs = incomeSubs.drop(['buckets'], axis=1)
incomeSubs.head(15)
```
